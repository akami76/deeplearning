```python
import platform
from matplotlib import font_manager, rc
import pandas as pd
cctv_seoul = pd.read_csv("cctv.csv", encoding='utf-8')
cctv_seoul.rename(columns={cctv_seoul.columns[0]:'dist'}, inplace=True)
cctv_seoul.rename(columns={cctv_seoul.columns[1]:'_sum'}, inplace=True)
cctv_seoul['last_increase'] = (cctv_seoul['2016년']+cctv_seoul['2015년']+cctv_seoul['2014년'])/cctv_seoul['2013년도 이전']*100
cctv_seoul
```

```python
cctv_seoul.set_index('dist', inplace=True)
cctv_seoul.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
cctv_seoul['_sum'].sort_values().plot(kind="barh", grid=True, figsize=(10,10))
plt.title('서울 cctv 구별 설치 현황')
plt.xlabel('cctv 수')
plt.ylabel('구')
plt.show()
```

```python
cctv_seoul['last_increase'].sort_values().plot(kind="barh", grid=True, figsize=(10,10))
plt.title('서울 cctv 구별 최근 3년간 증가율')
plt.xlabel('cctv 수')
plt.ylabel('구')
plt.show()
```
