```python
#import pandas as pd
#naver_rank_df = pd.read_html("https://www.naver.com/index.html", index_col='주요재무정보')

import requests
from bs4 import BeautifulSoup

url ="https://www.naver.com/index.html"

response = requests.get(url)
html = response.text

#print(html)
 
#뷰티풀소프의 인자값 지정
soup = BeautifulSoup(html, 'html.parser')
 
#쪼개기
title_list = soup.select('.ah_item')

#print(title_list)

for tag in title_list:
    print(tag.text)
    
    
```
